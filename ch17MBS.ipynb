```python
# estimate cash flow for MBS
mrate=8/100
prate=7/100
pv=100000000
fv=0
mat=360
cpr=0.002*1.5*1

# step 1: estimate MBS interest payment
pt_int=prate*pv/12

# Step 2: Estimate scheduled mortgage payment in a specific month 
pmt=-npf.pmt(mrate/12, mat, pv, fv)

#Step 3: Estimate the scheduled principal payment
mort_int=pv*mrate/12  
schedulepay=pmt-mort_int

#Step 4: Calculate the mortgage monthly prepayment 
smm=1-pow((1-cpr),1/12)
prepay=smm*(pv-schedulepay)

#Step 5: Calculate the cash flow to MBS investors and the MBSâ€™ ending balance
pt_pay=schedulepay+prepay
pt_cashflow=pt_pay+pt_int
pv=pv-pt_pay

print("MBS interest payment in month 1 =", round(pt_int,2))
print("scheduled mortgage payment in month 1 =", round(pmt,2)) 
print("scheduled principal payment in month 1 =", round(schedulepay,2))
print("prepayment in month 1 =", round(prepay,2))
print("MBS cash flow in month 1=", round(pt_cashflow,2))
print("ending balance in month 1", round(pv,2))
```

```python
mrate=8/100
prate=7/100
pv=100000000
fv=0
# create a pandas DataFrame
dfs=pd.DataFrame({})   
for i in range(360):
    month=i+1
    mat=360-month+1
    if month <= 30:
        cpr=0.002*1.5*month
    else:
        cpr=0.06*1.5
    smm=1-pow((1-cpr),1/12)
    pmt=-npf.pmt(mrate/12, mat, pv, fv)
    mort_int=pv*mrate/12  
    schedulepay=pmt-mort_int
    prepay=smm*(pv-schedulepay)
    pt_pay=schedulepay+prepay
    pt_int=prate*pv/12
    pt_cashflow=pt_pay+pt_int
    pv=pv-pt_pay
    # add one row to the DataFrame
    df=pd.DataFrame({'month':[month], 
                    'smm':[round(smm,6)], 
                    'pmt':[round(pmt,2)], 
                    'mort_int':[round(mort_int,2)], 
                    'ipmt':[round(pt_int,2)], 
                    'ppmt':[round(schedulepay,2)], 
                    'prepay':[ round(prepay,2)], 
                    'pt_pay':[ round(pt_pay,2)], 
                    'MBS_CF':[ round(pt_cashflow,2)],                     
                    'pv':[round(pv,2)]})
    df=df.set_index('month')
    dfs=pd.concat([dfs,df],axis=0)    
print(dfs)
```
