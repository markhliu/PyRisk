# Chapter 12: Foreign Exchange Risk


This chapter discusses foreign exchange risks faced by modern corporations, which can be classified into three categories: transactional risks, economic risks, and translational risks. We will examine each type of foreign exchange risk and discuss practical methods for managing these risks using derivative contracts.

Chapter Objectives:

1. Define foreign exchange rate risk
2. Identify foreign exchange risks faced by modern corporations
3. Understand the relationship between foreign exchange risk and hedging strategies

```python
import numpy as np
import matplotlib.pyplot as plt

fx = np.linspace(0.9,1.5,1000)
payment = np.zeros((1000,))

for i in range(1000):
    payment[i] = -fx[i]

fig = plt.figure(dpi=200, figsize=(10,6))
axes = plt.gca()
axes.set_xlim([0.9,1.5])
axes.set_ylim([-1.5,-0.9])

x1 = np.array([0.9,1.2])
y1 = np.array([-1.2,-1.2])
x2 = np.array([1.2,1.2])
y2 = np.array([-1.5,-1.2])

plt.plot(x1, y1, '--',color='orange')
plt.plot(x2, y2, '--',color='orange')

plt.plot(fx, payment, color='blue', linewidth = '5')
plt.xlabel('Euro exchange rate ($/)',fontsize=12)
plt.ylabel('Debt payment (in billion dollars)',fontsize=12)
plt.title('Debt payment as a function of the exchange rate',fontsize=14)
plt.show()
```

```python
net = np.zeros((1000,))
for i in range(1000):
    if fx[i]>=1.2:
        net[i] = fx[i]-1.2-0.025
    else:
        net[i] = -0.025

fig = plt.figure(dpi=200, figsize=(10,6))
axes = plt.gca()
axes.set_xlim([0.9,1.5])
axes.set_ylim([-0.4,0.4])

x1 = np.array([0.9,1.2])
y1 = np.array([-0.025,-0.025])
x2 = np.array([1.2,1.2])
y2 = np.array([-0.4,-0.025])

plt.plot(x1, y1, '--',color='orange')
plt.plot(x2, y2, '--',color='orange')

plt.plot(fx, net, color='blue', linewidth = '5')
plt.xlabel('Euro exchange rate ($/)',fontsize=12)
plt.ylabel('Net payoff from options (in billion dollars)',fontsize=12)
plt.title('Net payoff from options as a function of the exchange rate',fontsize=14)
plt.show()
```

```python
combined = np.zeros((1000,))
for i in range(1000):
    combined[i]=payment[i]+net[i]

fig = plt.figure(dpi=200, figsize=(10,6))
axes = plt.gca()
axes.set_xlim([0.9,1.5])
axes.set_ylim([-1.5,-0.9])

x1 = np.array([0.9,1.2])
y1 = np.array([-1.225,-1.225])
x2 = np.array([1.2,1.2])
y2 = np.array([-1.5,-1.225])

plt.plot(x1, y1, '--',color='orange')
plt.plot(x2, y2, '--',color='orange')

plt.plot(fx, combined, color='blue', linewidth = '5')
plt.xlabel('Euro exchange rate ($/)',fontsize=12)
plt.ylabel('Combined payment (in billion dollars)',fontsize=12)
plt.title('Combined payment as a function of the exchange rate',fontsize=14)

plt.show()
```
