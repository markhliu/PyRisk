# Chapter 11: Foreign Exchange Risk

```python
import numpy as np
import matplotlib.pyplot as plt

fx = np.linspace(0.65,0.9,1000) 
payment = np.zeros((1000,))

for i in range(1000):
    payment[i] = -100*fx[i]

fig = plt.figure(dpi=128, figsize=(10,6))
axes = plt.gca()
axes.set_xlim([0.65,0.95])
axes.set_ylim([-60,-95])

x1 = np.array([0.65,0.8])
y1 = np.array([-80,-80])
x2 = np.array([0.8,0.8])
y2 = np.array([-60,-80])

plt.plot(x1, y1, '--',color='orange')
plt.plot(x2, y2, '--',color='orange')

plt.plot(fx, payment, color='blue', linewidth = '5')
plt.xlabel('Canadian dollar exchange rate ($/CAD)',fontsize=12)
plt.ylabel('total payment (in million dollars)',fontsize=12)
plt.title('Expected Payment As a Function of Exchange Rate',fontsize=14)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

fx = np.linspace(0.65,0.9,1000) 
payoff = np.zeros((1000,))
put_expense = 1.5

for i in range(1000):
    if fx[i]>0.8:
        payoff[i] = 100*(fx[i]-0.8)-put_expense
    else:
        payoff[i] = -put_expense

fig = plt.figure(dpi=128, figsize=(10,6))
axes = plt.gca()
axes.set_xlim([0.65,0.95])
axes.set_ylim([-4,10])

plt.plot(fx, payoff, color='orange', linewidth = '5')
plt.xlabel('Canadian dollar exchange rate ($/CAD)',fontsize=12)
plt.ylabel('total payoff (in million dollars)',fontsize=12)
plt.title('Total Payoff from Options As a Function of Exchange Rate',fontsize=14)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

fx = np.linspace(0.65,0.9,1000) 
combined = np.zeros((1000,))
call_expense = 1.5

for i in range(1000):
    if fx[i]>0.8:
        payoff[i] = -100*fx[i] + 100*(fx[i]-0.8)-call_expense
    else:
        payoff[i] = -100*fx[i] - call_expense

fig = plt.figure(dpi=128, figsize=(10,6))
axes = plt.gca()
axes.set_xlim([0.65,0.95])
axes.set_ylim([-90,-65])

x1 = np.array([0.65,0.8])
y1 = np.array([-81.5,-81.5])
x2 = np.array([0.8,0.8])
y2 = np.array([-81.5,-90])

plt.plot(x1, y1, '--',color='orange')
plt.plot(x2, y2, '--',color='orange')

plt.plot(fx, payoff, color='green', linewidth = '5')
plt.xlabel('Canadian dollar exchange rate ($/CAD)',fontsize=12)
plt.ylabel('Combined payment (in million dollars)',fontsize=12)
plt.title('Combined Payment As a Function of Exchange Rate',fontsize=14)

plt.show()
```
