# Chapter 8: Hedging with Financial Derivatives


Financial derivatives include instruments such as futures, forwards, options, and swaps. This chapter provides an introduction to basic tools, terminology, and concepts fundamental to financial risk management using financial derivatives. It also explores the specific applications of options, forwards, and swaps in hedging corporate risks.

Chapter Objectives:

1. Provide an overview of hedging activities

2. Discuss the use of options for hedging

3. Examine the use of forward contracts for hedging

4. Analyze the use of swaps for hedging.

5. Investigate the relationship between financial derivatives and insurance

```python
import numpy as np
import matplotlib.pyplot as plt

call_price = 4
ST = np.linspace(0,200,1000)
X = 120
CT = np.zeros((1000,))
CT_net = np.zeros((1000,))

for i in range(len(ST)):
    CT[i] = max(ST[i]-X,0)
    CT_net[i] = CT[i] - call_price
   
fig = plt.figure(dpi=200, figsize=(10,8))
axes = plt.gca()
axes.set_ylim([-10,30])
axes.set_xlim([100,150])
plt.plot(ST, CT, color='green', label='payoff at maturity',
         linestyle="dotted", linewidth = '5')
plt.plot(ST, CT_net, color='red', label='net payoff at maturity',
         linewidth = '5')
plt.legend()
plt.show()
```

```python
put_price = 2
ST = np.linspace(0,200,1000)
X = 65
PT = np.zeros((1000,))
PT_net = np.zeros((1000,))

for i in range(len(ST)):
    PT[i] = max(X-ST[i],0)
    PT_net[i] = PT[i] - put_price
   
fig = plt.figure(dpi=200, figsize=(10,8))
axes = plt.gca()
axes.set_ylim([-10,30])
axes.set_xlim([40,90])
plt.plot(ST, PT, color='blue', label='payoff at maturity',
         linestyle="dotted", linewidth = '5')
plt.plot(ST, PT_net, color='orange', label='net payoff at maturity',linewidth = '5')

plt.xlabel('Crude oil price at maturity',fontsize=12)
plt.ylabel('Payoff to the put option',fontsize=12)
plt.title('Payoff to the put option as a function of the crude oil price',fontsize=14)
plt.legend()
plt.show()
```

```python
price = np.linspace(4,6,1000)
payoff = np.zeros((1000,))
put_expense = 2
for i in range(1000):
    if price[i]<5:
        payoff[i] = 10*(5-price[i])-put_expense
    else:
        payoff[i] = -put_expense
fig = plt.figure(dpi=200, figsize=(10,6))
axes = plt.gca()
axes.set_xlim([3,7])
axes.set_ylim([-10,15])

x1 = np.array([3,3.5,4,4.5])
y1 = np.array([3,3,3,3])
x2 = np.array([4.5,4.5,4.5,4.5])
y2 = np.array([-10,-5,0,3])

plt.plot(x1, y1, '--',color='orange')
plt.plot(x2, y2, '--',color='orange')

plt.plot(price, payoff, color='blue', 
    label='payoff from the put options contracts',linewidth = '5')
plt.xlabel('Corn price per bushel (in dollars)',fontsize=12)
plt.ylabel('Net payoff (in million dollars)',fontsize=12)
plt.title('Net payoff from the put options contracts',fontsize=14)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# create 1000 values of the SP500 index
spvalue = np.linspace(3000,5000,1000)
units = 1000000/4000
# portfolio value
investments = units*spvalue
# payoff from teh futures contract
futures = 250 * (4000-spvalue)
# total combined value
total=investments+futures
# plot three values in 1 figure
fig = plt.figure(dpi=200, figsize=(10,6))
axes = plt.gca()
axes.set_xlim([3000,5000])
axes.set_ylim([-250000,1250000])
# plot portfolio value in a dashed line
plt.plot(spvalue, investments, color='blue', 
    label='value of the investment in S&P500',
    linewidth = '5',linestyle="dashed")
# payoff from futures contract in a dotted line
plt.plot(spvalue, futures, color='green', 
    label='payoff from the futures contract',
    linewidth = '5',linestyle="dotted")
# combined value in a solid line
plt.plot(spvalue, total, color='red', 
    label='total portfolio value',
    linewidth = '5')
plt.xlabel('Value of the S&P 500 index',fontsize=12)
plt.ylabel('Value (in million dollars)',fontsize=12)
plt.title('Portfolio value with futures hedging',fontsize=14)
plt.legend(loc="center right")
plt.grid()
plt.show()
```
