```python
import numpy as np
import matplotlib.pyplot as plt

call_price = 4
ST = np.linspace(0,200,1000) 
X = 120
CT = np.zeros((1000,))
CT_net = np.zeros((1000,))

for i in range(len(ST)):
    CT[i] = max(ST[i]-X,0)
    CT_net[i] = CT[i] - call_price
    
fig = plt.figure(dpi=128, figsize=(10,8))
axes = plt.gca()
axes.set_ylim([-10,30])
axes.set_xlim([100,150])
plt.plot(ST, CT, color='green', label='payoff at maturity',linewidth = '5')
plt.plot(ST, CT_net, color='red', label='net payoff at maturity',linewidth = '5')

plt.xlabel('IMB Stock Price at Maturity',fontsize=12)
plt.ylabel('Payoff to the Call Option',fontsize=12)
plt.title('IBM Stock Call Option Payoff As a Function of Stock Price',fontsize=14)
plt.legend()
plt.show()   
```

```python
import numpy as np
import matplotlib.pyplot as plt

call_price = 4
ST = np.linspace(0,200,1000) 
X = 120
CT = np.zeros((1000,))
CT_net = np.zeros((1000,))

for i in range(len(ST)):
    CT[i] = - max(ST[i]-X,0)
    CT_net[i] = CT[i] + call_price
    
fig = plt.figure(dpi=128, figsize=(10,8))
axes = plt.gca()
axes.set_ylim([-30,10])
axes.set_xlim([100,150])
plt.plot(ST, CT, color='green', label='payoff at maturity',linewidth = '5')
plt.plot(ST, CT_net, color='red', label='net payoff at maturity',linewidth = '5')

plt.xlabel('IMB Stock Price at Maturity',fontsize=12)
plt.ylabel('Payoff to the Call Option Seller',fontsize=12)
plt.title('Payoff to the IBM Stock Call Option Seller As a Function of Stock Price',fontsize=14)
plt.legend()
plt.show() 
```

```python
import numpy as np
import matplotlib.pyplot as plt

price = np.linspace(4,6,1000) 
payoff = np.zeros((1000,))
put_expense = 2

for i in range(1000):
    if price[i]<5:
        payoff[i] = 10*(5-price[i])-put_expense
    else:
        payoff[i] = -put_expense

fig = plt.figure(dpi=128, figsize=(10,6))
axes = plt.gca()
axes.set_xlim([3,7])
axes.set_ylim([-10,15])

x1 = np.array([3,3.5,4,4.5])
y1 = np.array([3,3,3,3])
x2 = np.array([4.5,4.5,4.5,4.5])
y2 = np.array([-10,-5,0,3])

plt.plot(x1, y1, '--',color='orange')
plt.plot(x2, y2, '--',color='orange')

plt.plot(price, payoff, color='blue', label='payoff from the put options contracts',linewidth = '5')
plt.xlabel('corn price per bushel (in dollars)',fontsize=12)
plt.ylabel('net payoff (in million dollars)',fontsize=12)
plt.title('Net Payoff from the Put Options Contracts',fontsize=14)

plt.show()
```
